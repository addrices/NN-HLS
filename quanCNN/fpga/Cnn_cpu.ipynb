```python
def Conv(inChannel,outChannel,size,img,kernel,bias):
    out = np.zeros((outChannel,size-2,size-2)) 
#     out = out.to(torch.int16)
    for p in range(outChannel):
        for i in range(size-2):
            for j in range(size-2):
                out[p][i][j] = bias[p]
    
    for q in range(inChannel):
        for p in range(outChannel):
            for i in range(size-2):
                for j in range(size-2):
                    for m in range (3):
                        for n in range(3):
                            out[p][i][j] = out[p][i][j] + kernel[p][q][m][n]*img[q][i+m][j+n]
#                             if( InChannel == 64 and i == 0 and j == 0 and p == 0):
#                                 print(img[q][i+m][j+n].item(),kernel[p][q][m][n].item(),out[p][i][j].item())
    return out    
    
def Pooling(Channel,size,img):
    print(size)
    out = np.zeros((Channel,int(size),int(size)))
    for i in range(Channel):
        for j in range(size):
            for k in range(size):
                out[i][j][k] = max(img[i][j*2][k*2],img[i][j*2+1][k*2],img[i][j*2][k*2+1],img[i][j*2+1][k*2+1])
    return out

def Pooling_3(Channel,size,img):
    print(size)
    out = np.zeros((Channel,int(size),int(size)))
    for i in range(Channel):
        for j in range(size):
            for k in range(size):
                out[i][j][k] = max(img[i][j*3][k*3],img[i][j*3+1][k*3],img[i][j*3+2][k*3],img[i][j*3][k*3+1],img[i][j*3+1][k*3+1],img[i][j*3+2][k*3+1],img[i][j*3][k*3+2],img[i][j*3+1][k*3+2],img[i][j*3+2][k*3+2])
    return out
```
